# ¿Puedes generar un código que tenga condiciones periódicas en la caja al calcular la función de correlación si la distancia máxima de interés es un décimo del tamaño de la caja? Hazlo sólo para los datos 1a y 1d.

```python
# Aqui cargamos las liberias a usar
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import scipy.special
import numpy as np
import math
```

### Generamos los conjuntos de puntos que se necesitan para el ejercicio

Funcion para construir caja

```python
def construyecaja (l, s):
    lo = (l - s / 2) - s / 2
    nxlado = int(lo / s) + 2
    px = []
    py = []
    c = 1
    for i in range(1, nxlado):
        for j in range(1, nxlado):
            px.append(s * i - s / 2)
            py.append(s * j - s / 2)
            c += 1
    return px, py
```

Para generar una caja periodica vamos a crear 8 cajas alrededor de la caja principal(4 en cada lado y 4 en las esquinas), estas cajas contienen una copia de los puntos de la caja 1. Despues vamos a filtrar los puntos de tal forma que unicamente se conserven los puntos ' periodicos' que tienen una distancia menor o igual a la maxima con respecto a los puntos de la caja original

```python
def periodica(x1, y1, lcaja, frac):
    x1 = np.asarray(x1)
    y1 = np.asarray(y1)
    
    ax = x1 - lcaja
    ay = y1 + lcaja
    
    bx = x1
    by = y1 + lcaja
    
    cx = x1 + lcaja
    cy = y1 + lcaja
    
    dx = x1 - lcaja
    dy = y1
    
    ex = x1 + lcaja
    ey = y1
    
    fx = x1 - lcaja
    fy = y1 - lcaja
    
    gx = x1
    gy = y1 - lcaja
    
    hx = x1 + lcaja
    hy = y1 - lcaja
    
    px = []
    py = []
    x = []
    y = []
    
    px = np.block([x1, ax, bx, cx, dx, ex, fx, gx, hx])
    py = np.block([y1, ay, by, cy, dy, ey, fy, gy, hy])

    q1 = np.linspace(0, 0, 2)
    q2 = np.linspace(0, longitud, 2)
    q3 = np.linspace(longitud, longitud, 2)
    
    for i in range(len(px)):
        if px[i] > -lcaja * frac and px[i] < lcaja * frac + lcaja:
            if py[i] > -lcaja * frac and py[i] < lcaja * frac + lcaja:
                x.append(px[i]+ lcaja * frac)
                y.append(py[i]+ lcaja * frac)
                   
    plt.figure(figsize=(15, 15))
    plt.plot(q1 + lcaja * frac, q2 + lcaja * frac, color = 'g')
    plt.plot(q3 + lcaja * frac, q2 + lcaja * frac, color = 'g')
    plt.plot(q2 + lcaja * frac, q3 + lcaja * frac, color = 'g')
    plt.plot(q2 + lcaja * frac, q1 + lcaja * frac, color = 'g')
    plt.plot(x, y, "bo")
    return x, y
```

Funcion para integrar las funciones de correlacion.

```python
def integra(expresion, lcaja, nb):
    lmax = np.sqrt(2 * lcaja ** 2)
    dx = lmax / nb
    u = []
    for j in range(0, len(expresion)-1):
        are = np.abs((expresion[j] + expresion[j+1])/2 * dx)
        u.append(are)
    return u
```

Funciones para construir anillos de forma aleatoria dentro de la caja.

```python
def construyearo(r, n):
    xx = []
    yy = []
    da = 360/n
    for i in range(0, n):
        fx = r * math.cos((i * da)*np.pi/180)
        fy = r * math.sin((i * da)*np.pi/180)
        xx.append(fx)
        yy.append(fy) 
    return xx, yy

def esparceanillos(nanillos, longcaja, radioaros, pxanillo):
    px, py = construyearo(radioaros, pxanillo)
    x = []
    y = []
    for i in range(0, nanillos):
        centrox = np.random.uniform(-longcaja/2+radioaros, longcaja/2-radioaros)
        centroy = np.random.uniform(-longcaja/2+radioaros, longcaja/2-radioaros)
        for j in range(0, pxanillo):
            mx = px[j] + centrox
            my = py[j] + centroy
            if mx > -longcaja/2 and mx < longcaja/2:                
                if my > -longcaja/2 and my < longcaja/2:
                    x.append(mx+longcaja/2)
                    y.append(my+longcaja/2)
    return x, y
```

La funcion para obtener la distancia entre los puntos(sin repetir) se modifica para calcular unicamente la distancia ente los puntos dentro de la caja original (sin condiciones periodicas)

```python
def distancia(xw1, yw1, lcaja, frac):
    d = []
    i=0
    while i <= (len(xw1) - 1):
        j = (i + 1)
        while j < len(xw1):
            if xw1[i] > lcaja * frac and xw1[i] < lcaja * frac + lcaja:
                if yw1[i] > lcaja * frac and yw1[i] < lcaja * frac + lcaja:
                    if pow(((xw1[i] - xw1[j]) ** 2 + (yw1[i] - yw1[j]) ** 2), 0.5) <= lcaja * frac:
                        d.append(pow(((xw1[i] - xw1[j]) ** 2 + (yw1[i] - yw1[j]) ** 2), 0.5))
            j = j + 1
        i= i + 1
    return d
```

Funcion para obtener la distancia entre los puntos de una distribucion aleatoria y una en forma de malla (sin repetir)

```python
def distanciadr(x1,y1, xr, yr, lcaja, frac):
    d = []
    i=0
    while i <= (len(x1) - 1):
        j = (i + 1)
        while j < len(x1):
            if x1[i] > lcaja * frac and x1[i] < lcaja * frac + lcaja:
                if y1[i] > lcaja * frac and y1[i] < lcaja * frac + lcaja:
                    if pow(((x1[i] - xr[j]) ** 2 + (y1[i] - yr[j]) ** 2), 0.5) <= lcaja * frac:
                        d.append(pow(((x1[i] - xr[j]) ** 2 + (y1[i] - yr[j]) ** 2), 0.5))
            j = j + 1
        i= i + 1
    return d
```

Funcion para la correlacion de Hamilton

```python
def fham(xm, ym, xr, yr, lcaja, nb, frac):
    lmax = np.sqrt(2 * lcaja ** 2)  # Longitud maxima que puede haber en la caja
    dmalla = []
    drandom = []
    co = []
    
    dmalla = distancia(xm, ym, lcaja, frac) # Calculamos la distancia entre los puntos del arreglo en malla
    drandom = distancia(xr, yr, lcaja, frac) # Calculamos la distancia entre los puntos del arreglo de puntos aleatorios
    ddr = distanciadr(xm, ym, xr, yr, lcaja, frac) # Calculamos la distancia entre los puntos ordenados y los aleatorios
    
    dd, nbm = np.histogram(dmalla, bins = nb)
    rr, nbr = np.histogram(drandom, bins = nb)
    dr, nbrd = np.histogram(ddr, bins = nb)
    
    co = np.abs(dd * rr) / (dr ** 2) - 1# Calculamos la funcion de correlacion entre los histogramas

    x = np.linspace(0, lmax, len(co))
    
    plt.figure()
    plt.figure(figsize=(15,5))
    plt.plot(x, co, 'bo-', label='$Hamilton$', color = 'r')
    plt.legend()
    
    ham = integra(co, lcaja, nb)
    
    x2 = np.linspace(0, lmax, len(ham))
    
    plt.figure()
    plt.figure(figsize=(15,5))
    plt.plot(x2, ham, 'bo-', label='$Integral$', color = 'r')
    plt.legend()
    print( "Valor de la integral de la funcion de correlacion: ", sum(ham))
    return co
```

Funcion para la correlacion de Peebles - Hauser

```python
def fpeha(xm, ym, xr, yr, lcaja, nb, frac):
    lmax = np.sqrt(2 * lcaja ** 2)  # Longitud maxima que puede haber en la caja
    nrd = len(xr)
    n = len(xm)
    dmalla = []
    drandom = []
    co = []
    
    dmalla = distancia(xm, ym, lcaja, frac) # Calculamos la distancia entre los puntos del arreglo en malla
    drandom = distancia(xr, yr, lcaja, frac) # Calculamos la distancia entre los puntos del arreglo de puntos aleatorios
   
    dd, nbm = np.histogram(dmalla, bins = nb)
    rr, nbr = np.histogram(drandom, bins = nb)
    
    co = (nrd / n) ** 2 * (dd / rr) - 1 # Calculamos la funcion de correlacion entre los histogramas

    x = np.linspace(0, lmax, len(co))

    plt.figure()
    plt.figure(figsize=(15,5))
    plt.plot(x, co, 'bo-', label='$Peebles - Hauser$', color = 'g')
    plt.legend()
    
    ph = integra(co, lcaja, nb)
    
    x2 = np.linspace(0, lmax, len(ph))
    
    plt.figure()
    plt.figure(figsize=(15,5))
    plt.plot(x2, ph, 'bo-', label='$Integral$', color = 'g')
    plt.legend()
    print( "Valor de la integral de la funcion de correlacion: ", sum(ph))
    return co
```

Funcion para la correlacion de Estimador de Davis - Peebles

```python
def fdape(xm, ym, xr, yr, lcaja, nb, frac):
    lmax = np.sqrt(2 * lcaja ** 2)  # Longitud maxima que puede haber en la caja
    nrd = len(xr)
    n = len(xm)
    dmalla = []
    drandom = []
    co = []
    
    dmalla = distancia(xm, ym, lcaja, frac) # Calculamos la distancia entre los puntos del arreglo en malla
    ddr = distanciadr(xm, ym, xr, yr, lcaja, frac) # Calculamos la distancia entre los puntos ordenados y los aleatorios
    
    dd, nbm = np.histogram(dmalla, bins = nb)
    dr, nbrd = np.histogram(ddr, bins = nb)
    
    co = (nrd / n) * (dd / dr) - 1 # Calculamos la funcion de correlacion entre los histogramas

    x = np.linspace(0, lmax, len(co))

    plt.figure()
    plt.figure(figsize=(15,5))
    plt.plot(x, co, 'bo-', label='$Davis - Peebles$', color = 'm')
    plt.legend()
    
    ph = integra(co, lcaja, nb)
    
    x2 = np.linspace(0, lmax, len(ph))
    
    plt.figure()
    plt.figure(figsize=(15,5))
    plt.plot(x2, ph, 'bo-', label='$Integral$', color = 'm')
    plt.legend()
    print( "Valor de la integral de la funcion de correlacion: ", sum(ph))
    return co
```

Funcion para la correlacion de Landy - Szalay

```python
def flasz(xm, ym, xr, yr, lcaja, nb, frac):
    lmax = np.sqrt(2 * lcaja ** 2)  # Longitud maxima que puede haber en la caja
    nrd = len(xr)
    n = len(xm)
    dmalla = []
    drandom = []
    co = []
    
    dmalla = distancia(xm, ym, lcaja, frac) # Calculamos la distancia entre los puntos del arreglo en malla
    drandom = distancia(xr, yr, lcaja, frac) # Calculamos la distancia entre los puntos del arreglo de puntos aleatorios
    ddr = distanciadr(xm, ym, xr, yr, lcaja, frac) # Calculamos la distancia entre los puntos ordenados y los aleatorios
    
    dd, nbm = np.histogram(dmalla, bins = nb)
    rr, nbr = np.histogram(drandom, bins = nb)
    dr, nbrd = np.histogram(ddr, bins = nb)
    
    co = 1 + (dd / rr) * ((nrd / n) ** 2) - 2 * (dr / rr) * (nrd / n) # Calculamos la funcion de correlacion entre los histogramas

    x = np.linspace(0, lmax, len(co))

    plt.figure()
    plt.figure(figsize=(15,5))
    plt.plot(x, co, 'bo-', label='$Landy - Szalay$', color = 'k')
    
    
    ph = integra(co, lcaja, nb)
    
    x2 = np.linspace(0, lmax, len(ph))
    
    plt.figure()
    plt.figure(figsize=(15,5))
    plt.plot(x2, ph, 'bo-', label='$Integral$', color = 'k')
    plt.legend()
    print( "Valor de la integral de la funcion de correlacion: ", sum(ph))
    return co
```
