# Código que calcule histograma de distancias para 2d y 3d de una distribución de puntos aleatorios, escogiendo el tamaño de los bins, y normalizada a que su area sea 1.

```python
# Aqui cargamos las liberias a usar
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import scipy.special
import numpy as np
import math
```

```python
# Definimos algunas constantes
nptos = 2500
long = 1000
```

## 2 dimensiones

```python
# Construimos pares de numeros aleatorios
x1 = np.random.uniform(0, long, nptos)
y1 = np.random.uniform(0, long, nptos)
```

```python
# Graficamos los pares generados
plt.rcParams["figure.figsize"] = (7,7)
plt.plot(x1,y1, "*")
plt.xlabel("EJE X")
plt.ylabel("EJE Y")
```

```python
# Calculamos la distancia entre los puntos, sin repetir ningun par.
def distancia(x1,y1):
    d = []
    i=0
    while i <= (len(x1) - 1):
        j = (i + 1)
        while j < nptos:
            d.append(pow(((x1[i] - x1[j]) ** 2 + (y1[i] - y1[j]) ** 2), 0.5))
            j = j + 1
        i= i + 1
    return d
```

```python
# Graficamos el histograma
histograma1 = distancia(x1, y1)
nbins = 80
plt.figure()
plt.hist(histograma1, bins=nbins, color = 'g' , alpha =0.5)
plt.title("Histograma")
plt.ylabel("Frecuencia") 
plt.xlabel("Distancia")
plt.show()
```

```python
# Graficamos el histrograma normalizado
results, edges = np.histogram(d, normed=True)
binWidth = edges[1] - edges[0]
plt.bar(edges[:-1], results*binWidth, binWidth)
```

## 3 dimensiones

```python
# Construimos los elementos de cada 'dimension' por separado
x2 = np.random.uniform(0, long, nptos)
y2 = np.random.uniform(0, long, nptos)
z2 = np.random.uniform(0, long, nptos)
```

```python
# Graficamos
fig = pyplot.figure()
ax = Axes3D(fig)          
ax.scatter(x2, y2, z2)
plt.ylim(-1,long,long)
plt.rcParams["figure.figsize"] = (7,7)
```

```python
# Calculamos la distancia entre los puntos, sin repetir ningun par.
d2 = []
i=0
while i <= (nptos - 1):
    j = (i + 1)
    while j < nptos:
        d2.append(pow(((x2[i] - x2[j]) ** 2 + (y2[i] - y2[j]) ** 2 + (z2[i] - z2[j]) ** 2), 0.5))
        j = j + 1
    i= i + 1
histograma2 = np.array(d2)
```

```python
# Graficamos el histograma
nbins = 100
plt.figure()
plt.hist(histograma2, bins=nbins, color = 'g' , alpha =0.5)
plt.title("Histograma")
plt.ylabel("Frecuencia") 
plt.xlabel("Distancia")
plt.show()
```

```python
# Graficamos el histrograma normalizado
results2, edges2 = np.histogram(d2, normed=True)
binWidth2 = edges2[1] - edges2[0]
plt.bar(edges2[:-1], results2*binWidth2, binWidth2)
```
